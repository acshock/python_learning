```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

All the data here was taken from quandl. Each file covers different date ranges. Some go by month, others go by year. The files are merged by similar dates (either by month or by year). All the columns needed to be renamed since they were automatically called 'Value' which isn't all that useful. Several graphs were made to see if there were any relationships. The only thing of note is the bump that occurs around the start of the recession in ~2008. You can see this in all the "home" data: home ownership rate, home price, home sales. Since the crash was primarily about subprime mortgages and a lot of homes went into foreclosure as a result, that bump makes a lot of sense. The market seems to have made a pretty quick recovery except for home sales which are still low. The GDP per capita and the unemployment rate were also affected. The unemployment rate, however, is usually fairly independent of any type of "home" data. The GDP doesn't seem to have taken that big of a hit; it tracks the best with the price of homes.
